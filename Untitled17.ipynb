```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the images
ir_image = cv2.imread('/content/IR1.png', cv2.IMREAD_GRAYSCALE)  # Infrared image
vis_image = cv2.imread('/content/VIS1.png', cv2.IMREAD_GRAYSCALE)  # Visible image

# Check if images are loaded properly
if ir_image is None or vis_image is None:
    raise ValueError("One of the images could not be loaded. Check the file paths.")

# Define the zoom factor (e.g., 2x zoom)
zoom_factor = 2.0

# Get the dimensions of the IR image
height, width = ir_image.shape

# Calculate new dimensions
new_width = int(width * zoom_factor)
new_height = int(height * zoom_factor)

# Zoom the IR image using cv2.resize
zoomed_ir_image = cv2.resize(ir_image, (new_width, new_height), interpolation=cv2.INTER_LINEAR)

# Crop the zoomed image to match the size of the original VIS image
start_x = (new_width - width) // 2
start_y = (new_height - height) // 2
cropped_zoomed_ir = zoomed_ir_image[start_y:start_y + height, start_x:start_x + width]

# Display the images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Zoomed Infrared Image")
plt.imshow(cropped_zoomed_ir, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Original Visible Image")
plt.imshow(vis_image, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()

# Save the images if needed
cv2.imwrite('zoomed_infrared_image.png', cropped_zoomed_ir)
cv2.imwrite('original_visible_image.png', vis_image)
```

```python
# Import libraries
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import torch
from models.superpoint import SuperPoint
from models.superglue import SuperGlue
from models.utils import frame2tensor, make_matching_plot_fast

# Load SuperPoint and SuperGlue Models
device = 'cuda' if torch.cuda.is_available() else 'cpu'

# SuperPoint and SuperGlue Configuration
superpoint_config = {
    'nms_radius': 4,
    'keypoint_threshold': 0.005,
    'max_keypoints': 1024,
}
superglue_config = {
    'weights': 'outdoor',
    'sinkhorn_iterations': 20,
    'match_threshold': 0.2,
}

superpoint = SuperPoint(superpoint_config).eval().to(device)
superglue = SuperGlue(superglue_config).eval().to(device)

# Paths to input images
zoomed_ir_path = '/content/zoomedIR/zoomed_IR1.png'
pseudo_ir_path = '/content/pseudo_IR_images/VIS1.png'

# Load images
zoomed_ir = cv2.imread(zoomed_ir_path, cv2.IMREAD_GRAYSCALE)
pseudo_ir = cv2.imread(pseudo_ir_path, cv2.IMREAD_GRAYSCALE)

# Ensure pseudo IR is converted to grayscale
if pseudo_ir is None or len(pseudo_ir.shape) == 3:
    pseudo_ir = cv2.cvtColor(cv2.imread(pseudo_ir_path), cv2.COLOR_BGR2GRAY)

# Preprocess images (optional)
zoomed_ir = cv2.equalizeHist(zoomed_ir)
pseudo_ir = cv2.equalizeHist(pseudo_ir)

# Convert images to tensors
zoomed_tensor = frame2tensor(zoomed_ir, device)
pseudo_tensor = frame2tensor(pseudo_ir, device)

# Step 1: Detect Keypoints and Descriptors using SuperPoint
with torch.no_grad():
    zoomed_features = superpoint({'image': zoomed_tensor})
    pseudo_features = superpoint({'image': pseudo_tensor})

# Step 2: Match Keypoints using SuperGlue
with torch.no_grad():
    input_superglue = {
        'keypoints0': zoomed_features['keypoints'][0].unsqueeze(0),
        'keypoints1': pseudo_features['keypoints'][0].unsqueeze(0),
        'descriptors0': zoomed_features['descriptors'][0].unsqueeze(0),
        'descriptors1': pseudo_features['descriptors'][0].unsqueeze(0),
        'image0': zoomed_tensor,
        'image1': pseudo_tensor,
    }
    superglue_output = superglue(input_superglue)

matches = superglue_output['matches0'][0].cpu().numpy()
keypoints_zoomed = zoomed_features['keypoints'][0].cpu().numpy()
keypoints_pseudo = pseudo_features['keypoints'][0].cpu().numpy()

# Filter matches
valid = matches > -1
matched_zoomed_points = keypoints_zoomed[valid]
matched_pseudo_points = keypoints_pseudo[matches[valid]]

# Step 3: Compute Homography
matrix, mask = cv2.findHomography(matched_zoomed_points, matched_pseudo_points, cv2.RANSAC, 5.0)

# Step 4: Warp the zoomed IR image
h, w = pseudo_ir.shape
registered_ir = cv2.warpPerspective(zoomed_ir, matrix, (w, h))

# Save and visualize the results
cv2.imwrite('/content/registered_IR.png', registered_ir)

plt.figure(figsize=(10, 7))
plt.imshow(registered_ir, cmap='gray')
plt.title('Registered IR')
plt.axis('off')
plt.show()

# Optional: Visualize Matches
fig = make_matching_plot_fast(
    zoomed_ir, pseudo_ir, keypoints_zoomed, keypoints_pseudo, matches, valid,
    path='/content/matched_keypoints.png'
)
plt.imshow(fig)
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Paths to input images
registered_ir_path = '/content/registered_IR.png'
visible_image_path = '/content/data/data/datasets/visible2infrared/trainA/VIS1.png'

# Load images
registered_ir = cv2.imread(registered_ir_path, cv2.IMREAD_GRAYSCALE)
visible_image = cv2.imread(visible_image_path, cv2.IMREAD_GRAYSCALE)

# Step 1: Feature Detection and Matching
sift = cv2.SIFT_create()  # Use SIFT for keypoint detection and descriptor extraction
keypoints_ir, descriptors_ir = sift.detectAndCompute(registered_ir, None)
keypoints_vis, descriptors_vis = sift.detectAndCompute(visible_image, None)

# Match features using BFMatcher with ratio test
bf = cv2.BFMatcher(cv2.NORM_L2)
matches = bf.knnMatch(descriptors_ir, descriptors_vis, k=2)

# Apply Lowe's ratio test
good_matches = []
for m, n in matches:
    if m.distance < 0.75 * n.distance:
        good_matches.append(m)

# Step 2: Extract matched keypoints
if len(good_matches) >= 4:  # Minimum matches required for homography
    src_pts = np.float32([keypoints_ir[m.queryIdx].pt for m in good_matches]).reshape(-1, 1, 2)
    dst_pts = np.float32([keypoints_vis[m.trainIdx].pt for m in good_matches]).reshape(-1, 1, 2)

    # Step 3: Compute Homography to align registered IR to the visible image
    matrix, mask = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC, 5.0)

    # Warp the registered IR to the visible image's coordinate space
    h, w = visible_image.shape
    aligned_ir = cv2.warpPerspective(registered_ir, matrix, (w, h))

    # Step 4: Fuse aligned IR and visible image
    # Normalize images for proper blending
    aligned_ir = cv2.normalize(aligned_ir, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
    visible_image = cv2.normalize(visible_image, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

    # Fuse images using weighted averaging
    fused_image = cv2.addWeighted(visible_image, 0.6, aligned_ir, 0.4, 0)

    # Save and visualize the fused image
    fused_image_path = '/content/fused_image.png'
    cv2.imwrite(fused_image_path, fused_image)

    # Display results
    plt.figure(figsize=(10, 7))
    plt.imshow(fused_image, cmap='gray')
    plt.title('Fused Image')
    plt.axis('off')
    plt.show()

    print(f"Fused image saved at {fused_image_path}")
else:
    print("Not enough good matches found to align the images.")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Paths to input images
registered_ir_path = '/content/registered_IR.png'
visible_image_path = '/content/data/data/datasets/visible2infrared/trainA/VIS1.png'

# Load images
registered_ir = cv2.imread(registered_ir_path, cv2.IMREAD_GRAYSCALE)
visible_image = cv2.imread(visible_image_path, cv2.IMREAD_GRAYSCALE)

# Step 1: Preprocess Images
# Apply histogram equalization to both images
registered_ir = cv2.equalizeHist(registered_ir)
visible_image = cv2.equalizeHist(visible_image)

# Optionally apply Gaussian blur to reduce noise
registered_ir = cv2.GaussianBlur(registered_ir, (5, 5), 0)
visible_image = cv2.GaussianBlur(visible_image, (5, 5), 0)

# Step 2: Feature Detection and Matching using ORB
orb = cv2.ORB_create(nfeatures=2000)  # Increase features for better matching
keypoints_ir, descriptors_ir = orb.detectAndCompute(registered_ir, None)
keypoints_vis, descriptors_vis = orb.detectAndCompute(visible_image, None)

# Match features using BFMatcher with Hamming distance
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=False)
matches = bf.knnMatch(descriptors_ir, descriptors_vis, k=2)

# Apply Lowe's ratio test
good_matches = []
for m, n in matches:
    if m.distance < 0.6 * n.distance:  # Adjust threshold if needed
        good_matches.append(m)

# Step 3: Check for sufficient matches
if len(good_matches) >= 4:  # Minimum matches required for homography
    # Extract matched points
    src_pts = np.float32([keypoints_ir[m.queryIdx].pt for m in good_matches]).reshape(-1, 1, 2)
    dst_pts = np.float32([keypoints_vis[m.trainIdx].pt for m in good_matches]).reshape(-1, 1, 2)

    # Step 4: Compute Homography
    matrix, mask = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC, 5.0)

    # Step 5: Warp the registered IR to align with the visible image
    h, w = visible_image.shape
    aligned_ir = cv2.warpPerspective(registered_ir, matrix, (w, h))

    # Step 6: Fuse the aligned IR and visible image
    # Normalize images for proper blending
    aligned_ir = cv2.normalize(aligned_ir, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
    visible_image = cv2.normalize(visible_image, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

    # Fuse images using weighted averaging
    fused_image = cv2.addWeighted(visible_image, 0.6, aligned_ir, 0.4, 0)

    # Save and visualize the fused image
    fused_image_path = '/content/fused_image.png'
    cv2.imwrite(fused_image_path, fused_image)

    # Display results
    plt.figure(figsize=(10, 7))
    plt.imshow(fused_image, cmap='gray')
    plt.title('Fused Image')
    plt.axis('off')
    plt.show()

    print(f"Fused image saved at {fused_image_path}")
else:
    print(f"Not enough good matches found to align the images. Found: {len(good_matches)}")
```
